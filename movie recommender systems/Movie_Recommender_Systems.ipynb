```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(subset=None,keep="first",inplace=True)
```

```python
def actors(x):
    a = data[['movie_title','imdb_score']][data['actor_1_name'] == x]
    b = data[['movie_title','imdb_score']][data['actor_2_name'] == x]
    c = data[['movie_title','imdb_score']][data['actor_3_name'] == x]
    a = a.append(b)
    a = a.append(c)
    a = a.sort_values(by = 'imdb_score', ascending = False)
    return a.head(15)
```

```python
actors('Christian Bale')
```

```python
data['movie_title'] = data['movie_title'].str.strip()
pivot_table = data.pivot_table(index='movie_title', values='imdb_score')
movie_similarity = cosine_similarity(pivot_table)
similarity_df = pd.DataFrame(movie_similarity, index=pivot_table.index, columns=pivot_table.index)
```

```python
def recommend_movies(movie_title, num_recommendations=5):
    similar_movies = similarity_df[movie_title].sort_values(ascending=False)
    similar_movies = similar_movies.drop(movie_title)  # Exclude the input movie
    recommendations = similar_movies.head(num_recommendations).index.tolist()
    return recommendations
```

```python
movie_to_recommend = 'Avatar'
recommended_movies = recommend_movies(movie_to_recommend)
print(f"Recommended movies for '{movie_to_recommend}':")
for idx, movie in enumerate(recommended_movies, start=1):
    print(f"{idx}. {movie}")
```

```python
movie_input = widgets.Text(value='Avatar', description='Enter Movie Title:')
display(movie_input)


def on_button_click(button):
    movie_to_recommend = movie_input.value
    recommended_movies = recommend_movies(movie_to_recommend)
    with output:
        clear_output()
        print(f"Recommended movies for '{movie_to_recommend}':")
        for idx, movie in enumerate(recommended_movies, start=1):
            print(f"{idx}. {movie}")


button = widgets.Button(description="Get Recommendations")
display(button)
button.on_click(on_button_click)
output = widgets.Output()
display(output)
```
