```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='r', label='Iris setosa')
plt.scatter(X[y == 2, 0], X[y == 2, 1], c='b', label='Iris virginica')
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='g', label='Iris versicolor')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]


sns.pairplot(iris_df, hue='species', diag_kind='hist')
```

```python
sns.boxplot(x='species', y='petal length (cm)', data=iris_df)
plt.title('Box Plot of Petal Length by Species')
```

```python
sns.countplot(x='species', data=iris_df)
plt.title('Species Counts in Iris Dataset')
```

```python
sns.barplot(x='species', y='petal length (cm)', data=iris_df, errorbar=None)
plt.title('Mean Petal Length by Species')
```

```python
g = sns.PairGrid(iris_df, hue='species')
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot, kde_kws={'color': 'k'})
g.map_lower(sns.kdeplot, cmap='Blues_d')
```

```python
sns.violinplot(x='species', y='petal length (cm)', hue='species', data=iris_df)
plt.title('Violin Plot of Petal Length by Species and Sepal Width')
```
