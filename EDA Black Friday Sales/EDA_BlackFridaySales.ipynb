###Libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Black_friday_sale.csv",index_col=0,header=0)
```

```python
df.head(100)
```

```python
df.tail()
```

```python
df.describe()
```

```python
df=df.dropna()
```

```python
df.isna()
```

```python
df["Stay_In_Current_City_Years"].replace("4+",4,inplace=True)
```

```python
df["Age"].unique()
```

```python
df["Age"].replace(['26-35', '46-50', '51-55', '36-45', '18-25', '55+', '0-17'],["50","35","45","25","55","60","17"],inplace=True)
```

```python
df.dtypes
```

```python
pd.to_numeric(df["Age"])
```

```python
df.drop_duplicates(subset=None,keep="first",inplace=True,ignore_index=False)
```

```python
df.corr()
```

###Data Visualisations

```python
a=sns.displot(df["Age"])
```

```python
sns.kdeplot(data=df,x=df["Product_Category_1"],bw_adjust=1,multiple="stack")
sns.kdeplot(data=df,x=df["Product_Category_2"],bw_adjust=1,multiple="stack")
sns.kdeplot(data=df,x=df["Product_Category_3"],bw_adjust=1,multiple="stack")
```

```python
sns.kdeplot(data=df,x="Occupation",y="Product_Category_1")
```

```python
sns.scatterplot(data=df,x="Product_Category_3",y="Product_Category_2",hue="Age",style="Gender")
```

```python
sns.jointplot(data=df,x="Product_Category_3",y="Product_Category_1",kind="kde",hue="Gender")
```

```python
sns.pairplot(df,kind="kde",hue="Age")
```

```python
ax=sns.barplot(x="Occupation",y="Age",data=df)
```

```python
subset_columns = ['Age', 'Occupation', 'Stay_In_Current_City_Years', 'Product_Category_1']
sns.pairplot(df[subset_columns], diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
sns.boxplot(x=df["Product_Category_1"],data=df)
```

```python
sns.pairplot(df[subset_columns + ['City_Category']], hue='City_Category')
plt.suptitle('Pairwise Scatter Plot with Hue (City_Category)', y=1.02)
plt.show()
```

```python
sns.boxplot(x=df["Product_Category_1"],y="Age",data=df)
```

```python
sns.violinplot(x="Product_Category_1",y="Occupation",data=df)
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
